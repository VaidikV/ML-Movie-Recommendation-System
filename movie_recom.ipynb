```python
import numpy as np
import pandas as pd
```

```python
# genres
# id
# keywords
# title
# overview
# cast 
# crew

# only keeping the necessary columns

movies = movies[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew', 'production_companies']]
```

```python
# extracting genres from dict

import ast

def convert(genres):
    final_list = []
    for i in ast.literal_eval(genres):
        final_list.append(i['name'])
    return final_list
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'].apply(convert) # extracting keywords from dict
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
ast.literal_eval(movies['cast'][0])[0]
```

```python
# picking top 5 casts only and adding them to list

def convert_cast(cast_string):
    final_list = []
    counter = 0
    for i in ast.literal_eval(cast_string):
        if counter != 5:
            final_list.append(i['name'])
            counter += 1
        else:
            break
    return final_list
```

```python
movies['cast'].apply(convert_cast)
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
# fetching only directors from the crew column

def fetch_director(obj):
    final_list = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            final_list.append(i['name'])
            break
    return final_list
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'].apply(lambda x:x.split()) # converting overview to list
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# fetch production company name and convert to list

def fetch_production(obj):
    prodhouse = []
    try:
        prodhouse = [ast.literal_eval(obj)[0]['name']]
    except:
        prodhouse = []
    return prodhouse
```

```python
movies['production_companies'].apply(fetch_production)
```

```python
movies['production_companies'] = movies['production_companies'].apply(fetch_production)
```

```python
movies.head()
```

```python
# removing spaces (Science Fiction -> ScienceFiction)

movies['production_companies'] = movies['production_companies'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['production_companies']
```

```python
#stemming each word in tags

import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)

new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key=lambda x:x[1])[1:6]
# here we are using lambda to sort based on the second item i.e similarity distance and not the index which
# enumerate fn gave us.
```

```python
type(new_df)
```

```python
def recommend(movie):
    try:
        movie_index = new_df[new_df['title'] == movie].index[0]
    except IndexError:
        print("Movie Not Found...")
    else:
        distances = similarity[movie_index]
        movies_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:11]
    
        for i in movies_list:
            print(new_df.iloc[i[0]].title)
```

```python
recommend('Spirited Away')
```
